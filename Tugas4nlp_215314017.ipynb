```python
import spacy

# Muat model bahasa Inggris
nlp = spacy.load("en_core_web_sm")

# Contoh teks
text = "Elon Musk founded SpaceX in 2002 and Tesla in 2003."

# Proses teks dengan spaCy
doc = nlp(text)

# Tokenisasi
tokens = [token.text for token in doc]
print("Tokens:", tokens)

# Tampilkan entitas yang dikenali
for ent in doc.ents:
    print(ent.text, "→", ent.label_)
```

3. Implementasi NER dengan Hugging Face Transformers


```python
import stanza

# Download model NLP bahasa Indonesia
stanza.download("id")
nlp = stanza.Pipeline("id")

# Contoh teks dalam bahasa Indonesia
text = "Presiden mengunjungi negara itu pada tahun 2023."
doc = nlp(text)

# Cetak hasil dependency parsing
for sentence in doc.sentences:
    for word in sentence.words:
        print(f"{word.text} → {word.deprel} (head: {sentence.words[word.head - 1].text if word.head > 0 else 'ROOT'})")
```

2. Implementasi POS Tag dengan spaCy

```python
import spacy

# Load model bahasa Inggris
nlp = spacy.load("en_core_web_sm")

# Contoh teks
text = "Elon Musk founded SpaceX in 2002 and Tesla in 2003."

# POS Tagging
doc = nlp(text)

# Cetak hasil POS tagging
for token in doc:
    print(f"{token.text} → {token.pos_}")
```
